```python
## This code is for testing our model "model.h5" with live or existed videos
## In this example, we test it with a youtube car accident video

import cv2
import numpy as np
import os
import tensorflow as tf
import time
import pygame
```

```python
# load the saved model
model = tf.keras.models.load_model('model_new_approach.h5')
```

```python
# Define the list of class names
class_names = ['Abuse','Arrest','Arson','Assault','Burglary','Explosion','Fighting',"Normal",'RoadAccidents','Robbery','Shooting','Shoplifting','Stealing','Vandalism']

# Define the output directory path
output_dir_path = os.path.dirname('.')
```

```python
# Initialize pygame
pygame.init()

# Create a VideoCapture object
capture = cv2.VideoCapture('normal1.mp4')

# Set the initial width and height
width, height = 1600, 1700
capture.set(3, width)
capture.set(4, height) 

# Variable to control the blink rate
blink_rate = 0.5

# Load the alarm sound
pygame.mixer.music.load('alarm_sound.wav')

# Variable to track danger detection
danger_detected = False

frame_count= 0
state = False

# Function to apply the red filter
def apply_red_filter(frame):
    # Create a red filter by setting blue and green channels to 0
    red_filter = frame.copy()
    red_filter[:, :, 0] = 0
    red_filter[:, :, 1] = 0
    return red_filter
```

```python
# Loop through the video frames
while True:
    
    # Read a frame from the video
    ret, frame = capture.read()
    if not ret:
        break
    
    # Resize the frame to 64x64 pixels
    resized_frame = cv2.resize(frame, (64, 64))
        
    ## This section is optional, if you don't want to save the frames and have just the results
    ## you delete the 3 following code lines

    # Define the output image file path
    image_name = f'{frame_count}'
    output_image_path = os.path.join(output_dir_path, f"images/{image_name}.png")

    # Save the resized frame as a PNG image
    cv2.imwrite(output_image_path, resized_frame)

    # Apply the model to the resized frame
    input_data = np.expand_dims(resized_frame, axis=0)
    output_data = model.predict(input_data)

    # Map the output predictions to class names
    class_index = np.argmax(output_data)
    class_name = class_names[class_index]

    # Print the class name
    print(class_name)
    
    if class_name != 'normal':

        # Apply the red filter
        filtered_frame = apply_red_filter(frame)

        # Blink the red filter on and off
        if time.time() % (2 * blink_rate) < blink_rate:
            filtered_frame = frame
            state = False
        else:
            state = True

        # Play alarm sound if danger is detected

        if state:
            pygame.mixer.music.play()

        # Display the frame
        cv2.imshow('frame', filtered_frame)
    
    else:
        cv2.imshow('Camera', frame)
    
    if cv2.waitKey(10) & 0xFF==ord('d'):#20 milliseconds for every capture and when 
        break

# Release the VideoCapture object and close the windows
capture.release()
cv2.destroyAllWindows()

# Quit pygame
pygame.quit()
```

```python
import cv2
import time
import pygame

# Initialize pygame
pygame.init()

# Create a VideoCapture object
capture = cv2.VideoCapture(0)

# Set the initial width and height
width, height = 640, 480
capture.set(3, width)
capture.set(4, height)

# Variable to control the blink rate
blink_rate = 0.5

# Load the alarm sound
pygame.mixer.music.load('alarm_sound.wav')

# Variable to track danger detection
danger_detected = False

# Frame counter
frame_count = 0

# Function to apply the red filter
def apply_red_filter(frame):
    # Create a red filter by setting blue and green channels to 0
    red_filter = frame.copy()
    red_filter[:, :, 0] = 0
    red_filter[:, :, 1] = 0
    return red_filter

# Main loop
while True:
    # Read a frame from the video
    ret, frame = capture.read()

    # Check if the video has reached the end
    if not ret:
        break

    # Resize the frame to 64x64 pixels
    resized_frame = cv2.resize(frame, (64, 64))

    # Apply the model to the resized frame
    input_data = np.expand_dims(resized_frame, axis=0)
    output_data = model.predict(input_data)

    # Map the output predictions to class names
    class_index = np.argmax(output_data)
    class_name = class_names[class_index]

    # Print the class name
    print(class_name)

    if class_name != 'normal':
        # Apply the red filter
        filtered_frame = apply_red_filter(frame)

        # Blink the red filter on and off
        if time.time() % (2 * blink_rate) < blink_rate:
            filtered_frame = frame
            state = False
        else:
            danger_detected = False

        # Play alarm sound if danger is detected
        if danger_detected:
            pygame.mixer.music.play()

        # Display the frame
        cv2.imshow('frame', filtered_frame)
    else:
        cv2.imshow('Camera', frame)

    # Increment the frame count
    frame_count += 1

    # Wait for a small delay between frames (e.g., 10 milliseconds)
    if cv2.waitKey(10) & 0xFF == ord('d'):
        break

# Release the VideoCapture object and close the windows
capture.release()
cv2.destroyAllWindows()

# Quit pygame
pygame.quit()
```
